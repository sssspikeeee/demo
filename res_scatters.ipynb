```python
name = "DepCAGNN"
gt = np.load("./res/%s_gt.npy"%name)
preds = np.load("./res/%s_preds.npy"%name)

data_list = [gt, preds]
data = pd.DataFrame(data_list).transpose()
data.columns = ['GT', 'Pred']
scatter_plot(data, name)
```
