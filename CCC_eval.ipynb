```python
import numpy as np
import pickle
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# gt = np.load("./model_AAAI2023/res/gt_'AU_guided_0726'.npy")
# preds = np.load("./model_AAAI2023/res/preds_'AU_guided_0726'.npy")
# gt = np.load("./model_AAAI2023/res/gt_'AU_guided_0727_fusionGAT_residule'.npy")
# preds = np.load("./model_AAAI2023/res/preds_'AU_guided_0727_fusionGAT_residule'.npy")
# gt = np.load("./model_AAAI2023/res/gt_'AU_guided_0729_fusionGAT_residule'_cross.npy")
# preds = np.load("./model_AAAI2023/res/preds_'AU_guided_0729_fusionGAT_residule'_cross.npy")

gt1 = np.load("./model_AAAI2023/res/gt_'AU_guided_0728_northwind'.npy")
preds1 = np.load("./model_AAAI2023/res/preds_'AU_guided_0728_northwind'.npy")
gt2 = np.load("./model_AAAI2023/res/gt_'AU_guided_0728_freefrom'.npy")
preds2 = np.load("./model_AAAI2023/res/preds_'AU_guided_0728_freefrom'.npy")
gt = np.concatenate([gt1,gt2])
preds = np.concatenate([preds1,preds2])
```

```python
utt_len = [-1]
length = 0
for i in range(len(gt)-1):
    l = gt[i]
    l_next = gt[i+1]
    if l_next == l:
        length +=1
    else:
        utt_len.append(length)
        length = 0
```

```python
print(len(utt_len)-1)
print(utt_len)
```

```python
# gt_utt = gt[::32]
# pred_utt = preds[::32]

gt_utt = []
pred_utt = []

for i in range(len(utt_len)-1):
    gt_utt.append(utt_len[i]+1)
    # print(i)
    # print(utt_len[i]+1, utt_len[i+1])
    st, et = utt_len[i]+1, utt_len[i]+1+utt_len[i+1]
    # print(st, et)
    pred_utt.append(np.average(preds[st:et]))
```

```python
print(len(gt_utt))
print(len(pred_utt))
```

```python
def ccc_score(x, y):
    x_mean = np.nanmean(x)
    y_mean = np.nanmean(y)
    # x_mean = np.mean(x)
    # y_mean = np.mean(y)
    covariance = np.nanmean((x - x_mean) * (y - y_mean))
    x_var = np.nanmean((x - x_mean) ** 2)
    y_var = np.nanmean((y - y_mean) ** 2)
    CCC = (2 * covariance) / (x_var + y_var + (x_mean - y_mean) ** 2)
    
    return CCC
```

```python
ccc_score(gt_utt, pred_utt)
```

```python
ccc_score(gt, preds)
```

```python
rmse = np.sqrt(mean_squared_error(gt_utt, pred_utt))
print('RMSE : {:.4f}'.format(rmse))

rmse = np.sqrt(mean_squared_error(gt, preds))
print('RMSE : {:.4f}'.format(rmse))
```

```python
mae = mean_absolute_error(gt_utt, pred_utt)
print('MAE : {:.4f}'.format(mae))

mae = mean_absolute_error(gt, preds)
print('MAE : {:.4f}'.format(mae))
```
